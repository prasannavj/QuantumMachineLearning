```python
q0= cirq.GridQubit(0,0)
theta =sympy.Symbol('theta')
my_circuit =cirq.Circuit(cirq.ry(theta).on(q0))
```

```python
param_resolver ={theta:1.5}
state_vector = cirq.Simulator().simulate(my_circuit,param_resolver).final_state_vector
```

```python
z0=cirq.Z(q0)
z0.expectation_from_state_vector(state_vector,qubit_map={q0:0}).real
```
